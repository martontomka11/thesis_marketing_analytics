# Extract information from Podcast genre Y and Page X

Extract data from the 20 podcast shows on page X of genre Y 'Best podcasts'
e.g. https://www.listennotes.com/best-arts-podcasts-100/?region=us&sort_type=listen_score

```python
from listennotes import podcast_api

client = podcast_api.Client(api_key='insert key here')
response = client.fetch_best_podcasts(
  genre_id='68', # Change each extraction!
  page=10, # Change each extraction!
  region='us',
  sort='listen_score',
  safe_mode=0,
)
print(response.json())

podcasts = (response.json()['podcasts'])
```

Organize the variables and observations, write the results to a CSV file as a table:

```python
import pandas as pd 
import numpy as np
import csv

id = []
value = []
with open("podcasts_tvfilm_10.csv", "w") as csv_file: # Change each extraction!
    writer = csv.writer(csv_file, delimiter = ",")
    
    for i in podcasts:
        for k in i:
            id.append(k)
    writer.writerow(id[0:27])
    
    for i in podcasts:
        for k in i:
            value.append(str(i[k]).lstrip())    
        writer.writerow(value)
        print(value)
        value = []
```
